# Make an example histogram plot

Given some randomly distributed data, bin the data to create a histogram usuing maplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Create some random data

n = 500  #number of random data points to generate

x = np.random.randn(n)  #generate random data

#create the histogram bins
width = 0.5
histmin = np.floor(min(x))    #bottom of x, rounds down
histmax = np.ceil(max(x))+width    #top of x, rounds up

bins = np.arange(histmin,histmax,width)
plt.hist(x,bins=bins,alpha=0.5,edgecolor="black")    #plt.his makes and plots the histogram
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("example-histo.png",bbox_inches="tight",facecolor="white")
```
